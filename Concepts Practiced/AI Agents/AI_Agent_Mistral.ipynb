```python
from langchain_openai import ChatOpenAI

llm = ChatOpenAI(
    model="mistral-tiny",  # Can also try "mistral-small", "mistral-medium"
    temperature=0.7,
    openai_api_base="https://api.mistral.ai/v1",
    openai_api_key=os.environ["OPENAI_API_KEY"]
)
```

```python
# A Chat Test
from langchain.schema import HumanMessage

response = llm([HumanMessage(content="What is the difference between white chocolate and dark chocolate")])
print(response.content)
```
