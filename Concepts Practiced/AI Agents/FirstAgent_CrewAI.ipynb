```python
# define the agents

from crewai import Agent

researcher = Agent(
    role="AI Researcher",
    goal="Find top 3 recent AI trends",
    backstory="Expert at scanning tech blogs and news for latest trends.",
    allow_delegation=False
)

writer = Agent(
    role="Newsletter Writer",
    goal="Draft a clean, engaging article based on research",
    backstory="Writes in a friendly, newsletter style for tech audiences.",
    allow_delegation=False
)

editor = Agent(
    role="Editor",
    goal="Polish grammar and ensure clarity",
    backstory="Ensures perfect tone and readability of articles.",
    allow_delegation=False
)
```

```python
# Define the tasks

from crewai import Task

task1 = Task(
    description="Find and summarize 3 trending AI news stories from this week.",
    agent=researcher,
    expected_output="Bullet-point summary of 3 stories"
)

task2 = Task(
    description="Write a 400-word newsletter article explaining these stories.",
    agent=writer,
    expected_output="Formatted newsletter content"
)

task3 = Task(
    description="Polish the article for grammar, clarity, and structure.",
    agent=editor,
    expected_output="Final version ready for email or blog"
)
```

```python
# Create the crew and run it

from crewai import Crew

crew = Crew(
    agents=[researcher, writer, editor],
    tasks=[task1, task2, task3],
    verbose=True
)

final_output = crew.kickoff()
print("\n✅ Final Newsletter Article:\n", final_output)
```
